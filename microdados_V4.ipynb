```python
""" 
Importação das bibliotecas.
Carregamento dos dados e seleção das variáveis de interesse.
Mapeamento das variáveis categóricas para numéricas.
Tratamento de valores ausentes.
Padronização dos dados.
Aplicação do PCA antes da clusterização.
Determinação do número ideal de clusters usando o método do cotovelo e o coeficiente de Silhouette.
Aplicação do K-Means no conjunto de dados transformado pelo PCA.
Cálculo do coeficiente de Silhouette final para avaliar a qualidade da clusterização.
Visualização dos clusters no espaço bidimensional do PCA.
Análise das características médias de cada cluster.
"""
```

```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
microdados = pd.read_csv('microdados_tratado.csv', sep=';', encoding='ISO-8859-1')

"""
Leitura do arquivo JSON com os tipos de dados de cada coluna
Atualizar o arquivo JSON com os tipos de dados corretos e otimizados
"""
dtypes = pd.read_json("dtypes.json", typ='series')
microdados = microdados.astype(dtypes)

microdados = microdados.sample(100000, random_state=42)
```

```python
"""
Este bloco de código realiza uma análise de clusters sobre as variáveis de infraestrutura presentes no conjunto de dados `microdados_V4.ipynb`. A seguir, uma descrição detalhada das suas funcionalidades:

1. Seleção de Variáveis de Interesse:
   - As colunas de interesse relacionadas à infraestrutura, de `Q007` a `Q025`, são selecionadas e armazenadas na lista `colunas_infraestrutura`.

2. Criação de uma Cópia do DataFrame:
   - É criada uma cópia do DataFrame original contendo apenas as colunas de infraestrutura, garantindo que os dados originais não sejam alterados durante o processamento.

3. Mapeamento de Valores Categóricos para Numéricos:
   - Dois dicionários, `q7_q19` e `q18_q25`, são definidos para converter os valores categóricos das colunas selecionadas em valores numéricos.
   - As colunas de `Q007` a `Q017` e `Q019`, `Q022`, `Q024` são mapeadas usando `q7_q19`.
   - As colunas `Q018`, `Q020`, `Q021`, `Q023`, e `Q025` são mapeadas usando `q18_q25`.

4. Padronização dos Dados:
   - Os dados de infraestrutura são padronizados utilizando `StandardScaler` para normalizar as variáveis, o que é essencial para algoritmos de machine learning que são sensíveis à escala dos dados.

5. Redução de Dimensionalidade com PCA:
   - É aplicado o PCA (Análise de Componentes Principais) para reduzir a dimensionalidade dos dados padronizados para 2 componentes principais. Isso facilita a visualização e processamento posterior.

6. Aplicação do Algoritmo K-Means para Clusterização:
   - O algoritmo K-Means é utilizado para segmentar os dados em 3 clusters (k=3), com uma inicialização baseada em uma semente (`random_state=42`) para garantir a reprodutibilidade.
   - Os rótulos dos clusters resultantes são adicionados ao DataFrame original na coluna `cluster_infraestrutura`.

7. Atualização e Garantia da Tipagem Numérica:
   - As colunas de infraestrutura no DataFrame original são atualizadas com os valores mapeados e garantidas como numéricas, tratando possíveis erros com `pd.to_numeric` e `errors='coerce'`.

8. Cálculo do Coeficiente de Silhouette:
   - O coeficiente de Silhouette é calculado para avaliar a qualidade da clusterização, indicando o quão bem cada objeto está agrupado em seu cluster.

9. Visualização dos Clusters no Espaço PCA:
   - Utiliza-se `seaborn` para criar um gráfico de dispersão que visualiza os clusters no espaço das componentes principais, facilitando a interpretação visual da segmentação.

10. Análise das Características de Cada Cluster:
    - Calcula-se a média das variáveis de infraestrutura para cada cluster, adicionando uma coluna `Media_Geral` para facilitar a comparação entre os clusters.
    - Os clusters são então ordenados com base na `Media_Geral`.

11. Mapeamento dos Clusters para Categorias Descritivas:
    - Um dicionário `categoria_map` é criado para associar cada cluster a uma categoria descritiva, 
    como "Infraestrutura Básica Precária", "Infraestrutura Intermediária" e "Infraestrutura Completa ou Acima da Média".
    - Essa categorização facilita a interpretação dos clusters em termos mais compreensíveis.


"""
```

```python

# Selecionando as variáveis de interesse (Q007 a Q025)
colunas_infraestrutura = ['Q007', 'Q008', 'Q009', 'Q010', 'Q011', 'Q012',
                          'Q013', 'Q014', 'Q015', 'Q016', 'Q017', 'Q018',
                          'Q019', 'Q020', 'Q021', 'Q022', 'Q023', 'Q024',
                          'Q025']

# Criando uma cópia do DataFrame com as colunas de infraestrutura
infraestrutura = microdados[colunas_infraestrutura].copy()

# Dicionários de mapeamento completos
q7_q19 = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6}
q18_q25 = {'A': 0, 'B': 1}

# Aplicando o mapeamento para cada coluna correspondente
for col in ['Q007', 'Q008', 'Q009', 'Q010', 'Q011', 'Q012',
            'Q013', 'Q014', 'Q015', 'Q016', 'Q017', 'Q019',
            'Q022', 'Q024']:
    infraestrutura[col] = infraestrutura[col].map(q7_q19)

for col in ['Q018', 'Q020', 'Q021', 'Q023', 'Q025']:
    infraestrutura[col] = infraestrutura[col].map(q18_q25)

# Padronizando os dados
scaler = StandardScaler()
infraestrutura_scaled = scaler.fit_transform(infraestrutura)

# Aplicando PCA para reduzir a dimensionalidade
pca = PCA(n_components=2, random_state=42)
infraestrutura_pca = pca.fit_transform(infraestrutura_scaled)

# Aplicando o K-Means com 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(infraestrutura_pca)
microdados['cluster_infraestrutura'] = labels

# Atualizando as colunas de infraestrutura no microdados com os valores numéricos
microdados[colunas_infraestrutura] = infraestrutura[colunas_infraestrutura]

# Garantindo que as colunas de infraestrutura em microdados sejam numéricas
microdados[colunas_infraestrutura] = microdados[colunas_infraestrutura].apply(pd.to_numeric, errors='coerce')

# Calculando o coeficiente de Silhouette
score = silhouette_score(infraestrutura_pca, labels)
print(f'Coeficiente de Silhouette com K-Means para k=3: {score}')

# Visualizando os clusters no espaço PCA
plt.figure(figsize=(10, 7))
sns.scatterplot(x=infraestrutura_pca[:, 0], y=infraestrutura_pca[:, 1],
                hue=labels, palette='Set2')
plt.title('Clusters de Infraestrutura com K-Means (k=3)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Cluster')
plt.show()

# Analisando as características de cada cluster
cluster_summary = microdados.groupby('cluster_infraestrutura')[colunas_infraestrutura].mean()

# Calculando a média geral para cada cluster
cluster_summary['Media_Geral'] = cluster_summary.mean(axis=1)

# Ordenando os clusters pela média geral
cluster_summary = cluster_summary.sort_values('Media_Geral')

# Criando um dicionário para mapear os clusters às categorias
cluster_labels = cluster_summary.index.tolist()
categoria_map = {
    cluster_labels[0]: 'Infraestrutura Básica Precária',
    cluster_labels[1]: 'Infraestrutura Intermediária',
    cluster_labels[2]: 'Infraestrutura Completa ou Acima da Média'
}

# Adicionando a categoria ao DataFrame
microdados['categoria_infraestrutura'] = microdados['cluster_infraestrutura'].map(categoria_map)
```

```python
def plot_histograma(df, coluna_x, coluna_hue, dicionario_x=None, dicionario_hue=None, 
                   titulo='', rotacao_xticks=0, rotacao_xticks_barra=0,text_size=12, pad_titulo=20, palette='Set2'):
    """
    Plota um histograma de contagem com opcionalmente uma legenda (hue).
    """
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=coluna_x, hue=coluna_hue, palette=palette)

    # Adicionar valores no topo das barras
    for p in plt.gca().patches:
        height = p.get_height()
        plt.gca().text(p.get_x() + p.get_width()/2., height + 200, int(height), 
                      ha="center",rotation = rotacao_xticks_barra, size=text_size) 

    # Aplicar mapeamento aos rótulos do eixo x, se fornecido
    if dicionario_x:
        labels = [dicionario_x.get(x, x) for x in sorted(df[coluna_x].unique())]
        plt.gca().set_xticklabels(labels)
    
    # Aplicar mapeamento aos rótulos da legenda, se fornecido
    if dicionario_hue:
        handles, _ = plt.gca().get_legend_handles_labels()
        labels = [dicionario_hue.get(x, x) for x in sorted(df[coluna_hue].unique())]
        plt.legend(handles, labels, title=coluna_hue.replace('_', ' ').title())

    plt.title(titulo, pad=pad_titulo)
    plt.xlabel(coluna_x.replace('_', ' ').title())
    plt.ylabel('Quantidade')
    plt.xticks(rotation=rotacao_xticks)
    plt.legend(title=coluna_hue.replace('_', ' ').title(), loc= 'upper right')
    plt.tight_layout()
    plt.show()
```

```python
"""
Para calcular o desempenho, sera feito a média das notas das colunas:
NU_NOTA_COMP1
NU_NOTA_COMP2
NU_NOTA_COMP3
NU_NOTA_COMP4
NU_NOTA_COMP5
NU_NOTA_REDACAO

logo apos isso classificando em 4 categorias e criação de um dicionário para mapear as categorias:
- Muito Baixo: média < 300
- Baixo: 300 <= média < 500
- Médio: 500 <= média < 700
- Alto: média >= 700
"""

desempenho = {
    0: 'alto',
    1: 'médio',
    2: 'Baixo',
    3: 'Muito Baixo'
}

# Calculando a média das notas de redação e competências
microdados['NU_DESEMPENHO'] = microdados[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
                                            'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean(axis=1)

microdados['NU_DESEMPENHO'].unique()

# Função para mapear os valores
def map_desempenho(valor):
    if valor < 300:
        return 3
    elif 301 <= valor <= 500:
        return 2
    elif 501 <= valor <= 700:
        return 1
    else:
        return 0

# Aplicando a função à coluna 'categoria_desempenho'
microdados['NU_DESEMPENHO'] = microdados['NU_DESEMPENHO'].apply(map_desempenho)
```
